#### Copy files:

```python
import sys
import shutil

def copy_file(src_path, dst_path):
    try:
        shutil.copy(src_path, dst_path)
        message ='File Copied'
    except IOError as e:
        print(f'Unable to copy file {src_path} to {dst_path}')
        message ='Copy Failed - IO Error'
    except:
        print(f'When try copy file {src_path} to {dst_path}, unexpected error: {sys.exc_info()}')
        message ='Copy Failed - other Error'+ sys.exc_info()
        
    return message 
```

```python
df['Move_status'] = np.vectorize(copy_file)(df['image_path'], df['target_path'])
```

```python
df = df.drop(df[df['Move_status'].str.contains("Failed")].index, axis=0)
df
```

```python
import os

path = '.\\dataset - new\\classes\\covid\\'

for count, filename in enumerate(os.listdir(path)): 
        
        dst ="COVID19_" + str(count) + ".jpg"
        src = path + filename 
        dst = path + dst 
          
        os.rename(src, dst) 
```

```python
from os import walk

f = []

for (dirpath, dirnames, filenames) in walk(path):
    f.extend(filenames)
    break
```

```python
df['filename'] = f
df
```
